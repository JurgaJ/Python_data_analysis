```python
def gen_wave(Hz, sample_rate, length_sec, Amp, phase):
    t = np.linspace(0, length_sec, length_sec * sample_rate, endpoint=False)
    x = Amp*(np.sin(Hz * 2 * np.pi * t + phase))
    return(x, t)
```

```python
sample_rate = 1000
length_sec = 1
wave1, t = np.array(gen_wave(10, sample_rate, length_sec, 1, 0))
plt.plot(t, wave1)
plt.title('Wave')
plt.xlabel('Time, s')
plt.ylabel('Amplitude')
```

```python
wave3, t = np.array(gen_wave(10, sample_rate, length_sec, 3, 5))
plt.plot(t, wave3)
plt.title('Wave')
plt.xlabel('Time, s')
plt.ylabel('Amplitude')
```
